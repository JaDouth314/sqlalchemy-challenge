```python
import numpy as np
import pandas as pd
import datetime as dt
```

# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
# create engine to hawaii.sqlite
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# View all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Measurement = Base.classes.measurement
```

```python
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

# Exploratory Precipitation Analysis

```python
# Find the most recent date in the data set.
```

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results. 
# Starting from the most recent data point in the database. 

Latest_Date = (session.query(Measurement.date).\
               order_by(Measurement.date.desc()).\
               first())
print(Latest_Date)
```

```python
# Calculate the date one year from the last date in data set.

Year_From_Latest = dt.date(2017,8,23) - dt.timedelta(days=365)
print(Year_From_Latest)
```

```python
# Perform a query to retrieve the data and precipitation scores
Precipitation_scores = session.query(Measurement.date, Measurement.prcp).\
    filter(Measurement.date >= Year_From_Latest).\
    order_by(Measurement.date).all()

Precipitation_scores
```

```python
# Save the query results as a Pandas DataFrame and set the index to the date column
Precipitation_scores_df = pd.DataFrame(Precipitation_scores)
Precipitation_scores_df
```

```python
# Sort the dataframe by date
Sort_Precip_scores_df = Precipitation_scores_df.set_index('date')
Sort_Precip_scores_df
```

```python
# Use Pandas Plotting with Matplotlib to plot the data
df=pd.DataFrame(data=Precipitation_scores_df)
plt.figure(figsize = (14,6))
plt.plot(df['date'], df['prcp'])
plt.title("Precipitation in Hawaii (in Inches)")
plt.xlabel('Date')
# plt.xlim(0, max)
plt.ylabel('Precipitation')
plt.ylim(0,7)
plt.tight_layout()
plt.show()
```

```python
# Use Pandas to calcualte the summary statistics for the precipitation data
Precipitation_scores_df.describe()
```

# Exploratory Station Analysis

```python
# Design a query to calculate the total number stations in the dataset
Num_Stations = session.query(Station.station).count()
print(Num_Stations)
```

```python
# Design a query to find the most active stations (i.e. what stations have the most rows?)
# List the stations and the counts in descending order.
Station_counts = session.query(Measurement.station, func.count(Measurement.station)).\
    group_by(Measurement.station).\
    order_by(func.count(Measurement.station).desc()).all()

Station_counts
```

```python
Most_ActiveID = Station_counts[0][0]

Most_ActiveName = session.query(Station.name).filter_by(station = Most_ActiveID)
Most_ActiveName = Most_ActiveName[0][0]

print(f"The most active station is {Most_ActiveID}: {Most_ActiveName}")
```

```python
# Using the most active station id from the previous query, calculate the lowest, highest, and average temperature.
sel = [func.min(Measurement.tobs),
       func.max(Measurement.tobs),
       func.avg(Measurement.tobs)]
                
session.query(*sel).filter(Measurement.station=='USC00519281').all()
```

```python
# Using the most active station id
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram
StationYear = (session.query(Measurement.date, Measurement.tobs).\
                   filter(Measurement.date > Year_From_Latest).\
                   filter(Measurement.station == Most_ActiveID).\
                   order_by(Measurement.date).all())
StationYear_df = pd.DataFrame(StationYear)
StationYear_df = StationYear_df.set_index('date')
StationYear_df = StationYear_df.sort_index(ascending=True)
```

```python
StationYear_df.plot(kind='hist', bins=12)
plt.xlabel('Temperature')
plt.ylabel('Frequency')
```
